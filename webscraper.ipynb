```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://web.archive.org/web/20110514112442/http://unstats.un.org/unsd/demographic/products/socind/education.htm"

r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content)
```

```python
for row in soup('table'):
    print(row)
```

```python
soup('table')[6]
```

```python
print(soup('table')[6].get_text())
```

```python
print(soup('table')[6].get_text("|", strip=True))
```

```python
rows = soup('table')[6].findAll('tr')
print(rows)
```

```python
for row in rows:
    cells = row.findAll('td')
    
    output = []

    for i, cell in enumerate(cells):
        output.append(cell.text.strip())
    print(output)
```
